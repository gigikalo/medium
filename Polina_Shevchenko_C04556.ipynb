## T1.1: Describe a short pseudoalgorithm for finding i’th value in this data stream sequence. Discuss merits, restrictions and flaws.

I tried several approaches before ending up with the final one.

In particular, I tried to <b>bin the modulos on the fly</b> that would allow to store only the unique values and their count in a dictionary instead of all 8+ GB of 1bln integers. Once all 1bln integers are binned, it would be sufficient to sort the dictionary by keys for further operations. <br>
While the code was working and it was easy to work with the results, it was ridiculously slow, taking ca 1h to run on full volume of required data. 

<b>Example of such function bellow:</b>

```python

def randon_int_modulo(m,n):
    modulo = 2**m
    dict_ = {}
    random.seed(8) 
    dict_start_time = timer()
    i = 0
    while i <= n:
        value = np.random.randint(1, sys.maxsize)%modulo
        if value in dict_.keys():
            dict_[value] += 1
        else:
            dict_[value] = 1
        i+=1
    dict_time = timer() - dict_start_time
    return(dict_, m, n, dict_time)

```

I tried to speed it up by using multiprocessing which would allow me to 8 separate integers/dictionary generations and then I could combine the dictionaries. However, I didn't really succeed with this one either. 

I also though about an option to reduce the RAM usage by doing "kinda" multprocessing and spliting the generation,sorting, selection of the smallest integers into n loops. It would also improve the speed of sorting. But since we want to use the seed when generating the integers, the results for each loop would be the same - doesn't seem like what we want for this task.<br>
<b>Example of such loop below</b> (for 10mln random integers, searching for 10th smallest value):

```python

smallest_tens = []
i = 0
while i<100:
    seed = 1
    np.random.seed(seed)
    a = np.random.randint(0, sys.maxsize, 100000)%(2**20)
    a.sort()
    smallest_tens = smallest_tens + list(a[0:10])
    i+=1
```

After some other trials and fails <i>(like, using collections.Counter() instead of manual bining)</i>, I realised the most decent in terms of running time is, actually, the most dumb and memory inneficient solution - generating the whole 1 bln integers, run an in-memory sort and search<br>
The obvious drawbacks of this solution are memory related, it requires $O(n)$ amount of memory and hence cannot be scaled well. <br>
Sorting is also a heavy process, with average case time complexity $O(nlogn)$. <br>
Given all that, the implementation in Python is still much better than I could write myself for this exam. While this 'dumb'implementation is obviously not great, it ended up being the most suitable solution, given the amount of time and free memory I had.

````python
def order_statistics_results(modulo, seed_start, seed_end, n_int):
    time_result_table = pd.DataFrame()
    order_results_table = pd.DataFrame()
    
    time_results_columns = ['seed', 'generation time', 'sorting time', \
                            '100th i selection time', '100Kth i selection time', '10 mill i selection time',\
                            '100th i equal time', '100Kth i equal time','10 mill i equal time']
    order_results_columns = ['seed', '100th i', '100Kth i', '10 mill i',\
                             '100th i equal', '100Kth i equal', '10 mill i equal'] 
    
    for seed in range(seed_start,seed_end+1):
        print(seed)
        time_results = [seed]
        order_results = [seed]
        np.random.seed(seed)

        #generation
        start_time = timer()
        a = np.random.randint(1, modulo-1, n_int)
        int_generation_time = timer() - start_time
        time_results.append(int_generation_time)
        
#         print(int_generation_time)
        
        #sorting
        start_time = timer()
#         a = np.sort(a, kind='quicksort') #default; Edit: replaced with inplace
        a.sort()
        sorting_time = timer() - start_time
        time_results.append(sorting_time)

        #order int selection
        start_time = timer()
        a100_select = a[100]
        a100_select_time = timer() - start_time
        time_results.append(a100_select_time)
        order_results.append(a100_select)
        
#         print(a100_select_time)
        
        start_time = timer()
        a100000_select = a[100000]
        a100000_select_time = timer() - start_time
        time_results.append(a100000_select_time)
        order_results.append(a100000_select)
        
#         print(a100000_select_time)
        
        start_time = timer()
        a10000000_select = a[10000000]
        a10000000_select_time = timer() - start_time
        time_results.append(a10000000_select_time)
        order_results.append(a10000000_select)
        
#         print(a10000000_select_time)
        
        #count of equal ints
        start_time = timer()
        a100_equal_count = len(np.where(a == a100_select)[0])
        a100_equal_count_time = timer() - start_time
        time_results.append(a100_equal_count_time)
        order_results.append(a100_equal_count)        
        
        start_time = timer()
        a100000_equal_count = len(np.where(a == a100000_select)[0])
        a100000_equal_count_time = timer() - start_time
        time_results.append(a100000_equal_count_time)
        order_results.append(a100000_equal_count)
        
        start_time = timer()
        a10000000_equal_count = len(np.where(a == a10000000_select)[0])
        a10000000_equal_count_time = timer() - start_time
        time_results.append(a10000000_equal_count_time)
        order_results.append(a10000000_equal_count)
        
        del a
        time_result_table = time_result_table.append(pd.Series(time_results), ignore_index = True)
        order_results_table = order_results_table.append(pd.Series(order_results), ignore_index = True)
    
    time_result_table.columns = time_results_columns
    order_results_table.columns = order_results_columns
    return(time_result_table,order_results_table)
````

````python
def fibonacci(n, modulo):
    fibonacci = np.zeros(n)
    
    n1 = 1
    n2 = 0
    
    modulo_ = modulo-1
    for i in range(0, n):
        n_ = (n1 + n2) % (modulo_)
        
        fibonacci[i] = n_
        n2 = n1
        n1 = n_
        
    return(fibonacci)
````

````python
def Damerau_Levenshtein_dist (word_a, word_b):
    a = len(word_a)
    b = len(word_b)

    ab_matrix = pd.DataFrame(np.zeros((a+1,b+1)))
    # a - colums
    # b - rows

    ab_matrix.iloc[:,0] = [i for i in range(0,a+1)]
    ab_matrix.iloc[0,:] = [i for i in range(0,b+1)]

    for i in range(1,a+1):
        for j in range(1,b+1):
            ab_matrix.iloc[i,j] = np.min(\
               [ab_matrix.iloc[i-1,j-1] + (0 if word_a[i-1] == word_b[j-1] else\
               (0.5 if word_a[i-1]+word_b[j-1] in ['td','kg','bp','dt','gk','pb'] else 1)),
               ab_matrix.iloc[i,j-1] + 1,
               ab_matrix.iloc[i-1,j] + 1,   
               ab_matrix.iloc[i-2,j-2] + 1]
                        )
    return(ab_matrix.iloc[-1,-1])
````

````python
words = ['abbeystead','chiasm','chromoplasm','instead','masthead','mastheads','masturbate',\
         'masturbated','osteoaneurysm','spread']

edit_distances = pd.DataFrame(columns = words, index = words)

for word1 in words:
    for word2 in words:
        edit_distances.loc[word1,word2] = Damerau_Levenshtein_dist(word1,word2)
````

The alogirthm:
